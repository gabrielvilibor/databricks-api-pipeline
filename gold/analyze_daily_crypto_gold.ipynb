### Criação de uma view para analise aiária de Cryptos - Camada Gold

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import lag, col
```

```python
df = spark.table("workspace.silver.silver_crypto_prices")

window = Window.partitionBy("symbol").orderBy("ingestion_date")

df_gold = (
    df
    .withColumn("previous_price", lag("last").over(window))
    .withColumn("daily_variation", col("last") - col("previous_price"))
)

(
    df_gold
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("workspace.gold.gold_crypto_variation")
)
```
