### Ingest√£o de dados da tabela bronze na Silver com tratamento do tipo de dados(Python) - Camada Silver

```python
from pyspark.sql.functions import col, to_timestamp
```

```python
df = spark.table("workspace.bronze.bronze_crypto_prices")

df_silver = (
    df
    .withColumn("last", col("last").cast("double"))
    .withColumn("lowest", col("lowest").cast("double"))
    .withColumn("highest", col("highest").cast("double"))
    .withColumn("daily_change_percentage", col("daily_change_percentage").cast("double"))
    .withColumn("api_datetime", to_timestamp("api_datetime"))
    .dropDuplicates(["symbol", "ingestion_date"])
)

(
    df_silver
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("workspace.silver.silver_crypto_prices")
)
```
